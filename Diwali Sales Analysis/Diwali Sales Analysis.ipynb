# Exploratory Data Analysis

## Gender

```python
plt.figure(figsize=(4,5))
gen_cnt=sns.countplot(x='Gender',data=df)

for i in gen_cnt.containers:
    gen_cnt.bar_label(i)
    
plt.title('Gender Count')    
plt.show()
```

```python
plt.figure(figsize=(7,5))

gen_sales=df.groupby(by=['Gender'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
amnt=gen_sales['Amount']
l=[]
for i in gen_sales['Gender']:
    l.append(i)   

plt.pie(amnt,labels=l,autopct='%1.2f%%')
plt.title('Sales by Gender')

plt.show()
```

##### From the above graphs we can see that female buyers are more as compare to male buyers and purchasing power of women is also more than that of men.

## Age

```python
plt.figure(figsize=(9,5))

age_cnt=sns.countplot(x='Age Group',hue='Gender',data=df.sort_values(by=['Age Group']))

for i in age_cnt.containers:
    age_cnt.bar_label(i)

plt.title('Total Customers by Age Group')    
plt.show()
```

```python
plt.figure(figsize=(7,4))
age_grp_sales=df.groupby(by=['Age Group'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=age_grp_sales)

plt.title('Sales by Age Group')
plt.show()
```

##### From the above graph it is seen that most of the buyers are from Age Group of 26-35 yrs in which Female customers are significantly more than male customers

## State

```python
plt.figure(figsize=(8,5))

state_orders=df.groupby(by=['State'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False)
sns.barplot(y='State',x='Orders',data=state_orders)

plt.yticks(size=7)
plt.xlabel('Orders',size=10)
plt.ylabel('State',size=10)
plt.title('Orders by State',size=12)
plt.show()
```

```python
plt.figure(figsize=(15,5))

state_amount=df.groupby(by=['State'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
sns.barplot(x='State',y='Amount',data=state_amount)

plt.xticks(size=9)
plt.xlabel('State',size=13)
plt.ylabel('Amount',size=13)
plt.title('Sales by Top 10 State',size=15)
plt.show()
```

##### From the above graph it is seen that  the top performing states is terms of Orders and Sales/Amount are from Uttar Pradesh,Maharashtra and Karnataka where as Rajasthan,Punjab and Telangana are least in terms of Orders

## Marital Status

```python
plt.figure(figsize=(10,2))
tot_cnt=sns.countplot(y='Marital_Status',data=df)

for i in tot_cnt.containers:
    tot_cnt.bar_label(i)
    
plt.title('Marital Status')    
plt.show()
```

```python
plt.figure(figsize=(7,4))
tot_sales=df.groupby(by=['Marital_Status','Gender'],as_index=False)['Amount'].sum()
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=tot_sales)

plt.title('Sales by Marital Status')
plt.show()
```

##### From the above graph most of the buyers are married and in that women is more as compare to men and also they have high purchasing power

## Occupation

```python
plt.figure(figsize=(20,8))
tot_cnt=sns.countplot(x='Occupation',data=df)

for i in tot_cnt.containers:
    tot_cnt.bar_label(i) 
    
plt.xlabel('Occupation',size=13)
plt.ylabel('Count',size=13)
       
plt.show()
```

```python
occ_sales=df.groupby(by=['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(y='Occupation',x='Amount',data=occ_sales)
plt.show()
```

##### From the graph it is seen that, Most of the customers are from IT Sector, Healthcare,Aviation sectors

## Product Category

```python
plt.figure(figsize=(10,8))
tot_cnt=sns.countplot(y='Product_Category',data=df)

for i in tot_cnt.containers:
    tot_cnt.bar_label(i) 
    
plt.ylabel('Product Category',size=13)
plt.xlabel('Count',size=13)
       
plt.show()
```

```python
plt.figure(figsize=(20,8))
prod_sales=df.groupby(by=['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Product_Category',y='Amount',data=prod_sales)

plt.xticks(rotation=20)
plt.xlabel('Product Category',size=12)
plt.ylabel('Count',size=12)

plt.show()
```

##### From the above we can see that most of the orders are coming from Clothing & Apparel, Food, Electronics however maximum sales is coming from Food category

```python
plt.figure(figsize=(12,5))
top10=df.groupby(by=['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Amount',data=top10)

plt.title('Top 10 Products')

plt.show()
```

### Conclusion:

##### Married Women of age group 26-35 yrs from UP,Maharashtra,Karnataka working in IT,Healthcare and Aviation sector are more likely to buy products from Food,Clothing and Electronics Category
