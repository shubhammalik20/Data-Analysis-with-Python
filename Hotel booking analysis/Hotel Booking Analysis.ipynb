# Exploratory Data Analysis

```python
data=df['is_canceled'].value_counts(normalize=True)*100
data1=[round(i,2) for i in df['is_canceled'].value_counts(normalize=True)*100]
print(data)

plt.figure(figsize=(4,6))
plt.bar(['Not Cancelled','Cancelled'],data1,edgecolor='k')

for i, v in enumerate(data1):
    plt.annotate(str(v), xy=(i, v), ha='center', va='bottom')

plt.title('Reservation Status Count')
plt.ylabel('Percentage')
plt.xlabel('Reservation Status')
plt.show()
```

### The above graph shows the percentage of reservations that are cancelled to those are not. There are 37%  of reservations have been cancelled which has a significant impact on hotel earnings.

```python
plt.figure(figsize=(8,3))
sns.countplot(data=df,y='hotel',hue='is_canceled',palette='hls')
plt.title('Reservation Count by Different Hotels')
plt.legend(['Not Cancelled','Cancelled'])
plt.xlabel('Number of Reservations')
plt.ylabel('Hotel')
plt.show()
```

### From the above graph it is seen that City Hotels have more booking than Resort Hotel . It is possible that resort hotels are more expensive that City hotel. Although Resort hotels have less booking than City hotels but the amount of booking that are cancelled is also less in Resort Hotels.

#### Let's find out the percentage of bookings cancelled and not cancelled

```python
x=df.groupby(by=['hotel'],as_index=True)['is_canceled'].value_counts(normalize=True)*100
x
```

#### Hence in Resort Hotels have 72% booking that are not cancelled and 28% are cancelled where as in City hotel cancelled reservation percentage is 58% and not cancelled is 41%. Hence, City Hotels are mostly encountered with cancelled booking than Resort Hotels.

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
city_hotel=df[df['hotel']=='City Hotel']
```

```python
resort_hotel=resort_hotel.groupby(by=['reservation_status_date'])[['adr']].mean()
city_hotel=city_hotel.groupby(by=['reservation_status_date'])[['adr']].mean()
```

```python
plt.figure(figsize=(11,5))
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend()
plt.title('Average Daily Rate in City and Resort Hotels')
plt.show()
```

### The line graph shows that the average daily rate for a city hotel is less than the resort hotel.It is certain that on weekdays and holidays the price of Resort hotels may see a rise.

```python
df['Month']=df['reservation_status_date'].dt.month

plt.figure(figsize=(10,5))
sns.countplot(data=df,x='Month',hue='is_canceled')
plt.title('Reservation Status Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Reservation')
plt.legend(['Not Cancelled','Cancelled'])
plt.show()
```

### The grouped bar is used to analyze the months with the highest and lowest reservations levels according to reservation status, with respect to either the reservation is cancelled or not. As it can be seen, the most number of non cancelled reservations are in month of August and least in month of December. January is the month with the highest cancelled reservation and August is the least month in terms of cancelled reservation.

```python
cancel_data=df[df['is_canceled']==1]
adr_sum=cancel_data.groupby(by=['Month'])['adr'].sum().reset_index()
plt.figure(figsize=(10,5))
sns.barplot(data=adr_sum,x='Month',y='adr')
plt.title('ADR per Month')
plt.show()
```

### The bar graph demonstrate that cancellations are most common when the price is large and are least when the price is lower. There cost of accommodation is solely responsible for the cancellation. 

```python
top_10_country=cancel_data['country'].value_counts().head(10)

plt.figure(figsize=(7,7))
plt.pie(top_10_country.values,autopct='%.1f',labels=top_10_country.index)
plt.title('Top 10 Countries Cancellation Rate')
plt.show()
```

### From the pie char we have seen the top 10 countries where maximum reservations are cancelled, in that the top country is Portugal with highest number of cancellations.

#### Let's check the areas from where guests are visiting the hotels and making reservations. is it coming from Direct or Group,online or offline travel agents?

```python
cancel_data['market_segment'].value_counts(normalize=True)*100
```

```python
not_cancel_data=df[df['is_canceled']==0]
not_cancel_data['market_segment'].value_counts(normalize=True)*100
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='market_segment',hue='is_canceled')
plt.xlabel('Market Segment',size=12)
plt.ylabel('Reservation Count',size=12)
plt.legend(['Cancelled','Not Cancelled'])
plt.title('Reservation Status by Market Segment')
plt.show()
```

### From the above analysis we can see that around 47% visitors are coming from Online TA, 27% from Groups, 4% of clients have booked hotels directly where as least number of visitors are coming from Aviation segment.

### Maximum number of reservations cancelled are also coming from Online TA, then 21% from Offline TA/TO, 14% are from Direct and 4% cancelled from Groups.

```python
not_cancel_data=df[df['is_canceled']==0]
avg_adr_cancel=cancel_data.groupby(by=['reservation_status_date'])[['adr']].mean().sort_values(by='reservation_status_date').reset_index()
avg_adr_not_cancel=not_cancel_data.groupby(by=['reservation_status_date'])[['adr']].mean().sort_values(by='reservation_status_date').reset_index()

plt.figure(figsize=(15,7))
plt.plot(avg_adr_cancel['reservation_status_date'],avg_adr_cancel['adr'],label='Cancelled')
plt.plot(avg_adr_not_cancel['reservation_status_date'],avg_adr_not_cancel['adr'],label='Not-Cancelled')
plt.legend()
plt.title('Average Daily Rate',size=15)
plt.show()
```

### As seen from the above graph reservations are cancelled when the average daily price is higher and not  cancelled when it is lower
